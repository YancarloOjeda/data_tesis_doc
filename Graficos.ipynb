# Este código sirve para graficar las BD  

Se crearon cuatro BD, la primera es del entrenamiento, la segunda y tercera son de las pruebas de A y x respectivamente y la cuarta de las pruebas de A y x juntas. Con este código se grafica la primera y la cuarta.

```python
import pandas as pd
import numpy as np
from scipy import stats
import os
import re
import matplotlib.pyplot as plt
```

```python
ubicacion_BD = "/home/yan/datos_redes_tesis_doctorado/Con ITI/2-2(2)-2(1)-1 parcialmente/PRUEBAS_A_x.csv"

datos = pd.read_csv(ubicacion_BD, sep = ',', engine='python', keep_default_na=False, na_values=[' '])

df = pd.DataFrame(datos)
df
```

```python
x = np.arange(.99, 19)
plt.bar(x, df['A_x'], color = 'K', label=r'$Red-1$')
plt.legend(fontsize = 15)
plt.show()
```

```python
df.mean().plot(kind='barh')
```

```python
ubicacion_BD_x = "/home/yan/datos_redes_tesis_doctorado/Con ITI/2-2(2)-2(1)-1 parcialmente/datos_filtrados_pruebas_x/PROMEDIOS_total_.csv"

datos_x = pd.read_csv(ubicacion_BD_x, sep = ',', engine='python', keep_default_na=False, na_values=[' '])

df_x = pd.DataFrame(datos_x)
df_x
```

```python
df_x.mean().plot(kind='bar', color='k', alpha=.7)
```

```python
df_x_error = stats.sem(df_x, axis=0, ddof=0) 
labels = ['Ax', 'A', 'x', 'A(1)x', 'A(2)x', 'A(4)x']
x_pos = np.arange(len(labels))

Ax_mean = df_x['Ax'].mean()
A_mean = df_x['A'].mean()
x_mean = df_x['x'].mean()
A1x_mean = df_x['A(1)x'].mean()
A2x_mean = df_x['A(2)x'].mean()
A4x_mean = df_x['A(4)x'].mean()

medias = [Ax_mean, A_mean, x_mean, A1x_mean, A2x_mean, A4x_mean]
```

```python
fig, ax = plt.subplots()

ax.bar(x_pos, medias,
       yerr=df_x_error,
       align='center',
       alpha=.7,
       color='k',
       capsize=10)
ax.set_ylabel(r"Activación $M'$", fontsize = 20)
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
#ax.set_title('Resultados pruebas a x', fontsize = 20)
ax.set_xlabel('Grupos', fontsize = 20)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.yaxis.grid(False)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# Save the figure and show
plt.tight_layout()
plt.savefig('bar_plot_with_error_bars1.png', dpi=500)
plt.show()
```
